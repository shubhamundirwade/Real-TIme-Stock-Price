```python
# Importing GoogleNews library
from GoogleNews import GoogleNews
```

```python
# Set perios of today
news = GoogleNews(period='1d')

# Searching location wise and storing it into result variable
news.search("India")
result = news.result()
```

```python
# import Pandas library to save result file into dataframe
import pandas as pd
data = pd.DataFrame.from_dict(result)
data.head()
```

```python
# droping unnecesary img column

data = data.drop(columns=['img'])
data.head()
```
